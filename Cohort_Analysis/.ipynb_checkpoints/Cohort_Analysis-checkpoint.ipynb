# Cohort Analysis

A cohort analysis is a method where we group users according to some similarity, analyzing their behavior over time, helping us to find trends and monitor customer retention and sales. This kind of analysis can help us to understand our business health and the customers loyalty.

Also, considering that the product evolves, looking at groups over time can help us understand how these changes affects user behavior.

So we have a dataframe where each row refers to a user information, but we have several sparse columns that represents the months the user has used our app.

We initially need to transform this matrix into tidy data, we will also rename some columns to a simpler format.

Since we have monthly usage information, we will use monthly cohorts to continue the analysis. 

```python
#Renaming some columns to a simpler format
df = df.rename(columns={'Bought At - Year ':'Year', 'Bought At - Month': 'Month', 'Bought At - Week ': 'Week'})
```

```python
#Creating the cohort group with de date Y-m
df['CohortGroup'] = pd.to_datetime(df['Year'].map(str) + '-' + df['Month'].map(str), errors='coerce').apply(lambda x: x.strftime('%Y-%m'))
```

```python
#Replacing the 'app_used' information by 1
df = df.replace('app_used', 1)
```

```python
#Transforming the dataframe to tidy data format
cols_to_melt = df.columns[5:-1].to_list()
df = df.melt(id_vars=['User_IDS', 'Country', 'CohortGroup', 'Year', 'Month', 'Week'], value_vars=cols_to_melt, var_name='month_used', value_name='app_used')
cols_to_melt=[]
```

```python
#Sorting values by user id
df = df.sort_values('User_IDS')
df.head(10)
```

Since we intend to look at monthly behavior, we need to aggregate total users and use count by CohortGroup within the month wich the app was used (month_used).

```python
#Creating the variable month_used with Y-m structure
df['month_used'] = pd.to_datetime(df['month_used']).apply(lambda x: x.strftime('%Y-%m'))
```

```python
#Removing data of uses before user purchase
df = df[df['CohortGroup'] <= df['month_used']]
```

```python
def cohort_group(df):
  """Return a dataframe grouped by CohortGroup and month_used"""
  
  #Grouping dataframe
  grouped = df.groupby(['CohortGroup', 'month_used'])

  #Define aggregate step to other columns
  cohorts = grouped.agg({
      'User_IDS': pd.Series.nunique,
      'app_used': np.sum
  })

  #Change column name
  cohorts.rename(columns={'User_IDS': 'TotalUsers',
                         'app_used': 'Total_uses'}, inplace=True)
  
  return cohorts

cohorts = cohort_group(df)
```

```python
cohorts.head(15)
```

We want to see how each cohort has behaved in the months following their purchase, so we will index each group for the first month with available usage information after the purchase date.

```python
def create_cohort_period(df):
  """Creates a CohortPeriod column on the selected dataframe"""

  #Label for each cohort group
  
  def cohort_period(df):  
    df['CohortPeriod'] = np.arange(len(df)) +1
    return df

  df = df.groupby(level=0).apply(cohort_period)
  
  return df

cohorts = create_cohort_period(cohorts)
```

```python
cohorts.head(15)
```

```python
cohorts.reset_index(inplace=True)
cohorts.set_index(['CohortGroup', 'CohortPeriod'], inplace=True)

cohort_group_size = cohorts['TotalUsers'].groupby(level=0).first()
```

## Percentage cohort analysis

---



We will then make a matrix with the percentage of uses per group over time

```python
app_usage_percent = cohorts['Total_uses'].unstack(0).divide(cohort_group_size, axis=1)
app_usage_percent.head(10)
```

```python
def cohort_plot(df_percent, percent=True):
  """Lineplot and heatmap of the percent cohort analysis"""
  
  #Define subplots
  fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,8))
  
  
  df_percent.plot(ax=ax1)
  ax1.title.set_text('Cohorts: App Usage Lineplot')
  ax1.set_xticks(np.arange(1, 12.1, 1))
  ax2.title.set_text('Cohorts: App usage Heatmap')
  
  if percent == True:
    sns.heatmap(df_percent.T, mask=df_percent.T.isnull(), annot=True, ax=ax2)
    ax1.set_ylabel('% of users using the app')
  
  elif percent == False:
    sns.heatmap(df_percent.T, mask=df_percent.T.isnull(), annot=True, fmt='g', ax=ax2)
    ax1.set_ylabel('Number of users using the app')

  plt.show()
```

Plotting our matrix lineplot and heatmap.

```python
cohort_plot(app_usage_percent)
```

Surprisingly, we can see from the lineplot that user usage of the app has increased over time. About 5% of users in the first cohorts used our app at time 1, but usage increased to over 30% after 10 months.

Through the heatmap we realize that this increase occurs in diagonal layers, from the upper left to the lower right direction. This indicates that the increase in usage is related to the actual date, not the user's usage time. Some app improvements may have been made over time, leading to greater retention.

However, the last three cohorts do not appear to perform as well. It is hard to say what may have changed here, perhaps the user really needs some time to fit our meditation app into their daily routine.

## Absolute cohort analysis

Now we will use an dataframe with the absolute values of the number of uses of our application to plot the lineplot and the heatmap.

```python
app_usage_absolute = cohorts['Total_uses'].unstack(0)
```

```python
cohort_plot(app_usage_absolute, percent=False)
```

As we saw in the previous graphs, but now in absolute values, the number of users using our app has increased considerably. In group 2018-01 for example, the number of uses went from 17 to 189 after 10 months.

In the heatmap we still have very similar layers, but now with a slightly more top-down direction. The number of users has been increasing cohort after cohort, which indicates an increase in sales, since the percentages are not very different from each other. We see a spike in sales in the 2018-11 cohort, returning close to normal in subsequent cohorts.



## Cohort analysis by country

So let's take a quick look at the same graphs, but now filtering the data by country, just to make sure that they all have the same trend as we initially saw.

```python
df.head()
```

```python
# Countrys list
country_list = df['Country'].unique().tolist()

print(country_list)
```

```python
class color:
   BOLD = '\033[1m'
```

```python
def cohort_transform_plot(df):
  """Plot the cohort analysis from the specified dataframe"""
  
  #Group cohort and create cohort period var
  cohorts = cohort_group(df)
  cohorts = create_cohort_period(cohorts)

  #Rearange the cohorts dataframe index
  cohorts.reset_index(inplace=True)
  cohorts.set_index(['CohortGroup', 'CohortPeriod'], inplace=True)

  cohort_group_size = cohorts['TotalUsers'].groupby(level=0).first()
  app_usage_percent = cohorts['Total_uses'].unstack(0).divide(cohort_group_size, axis=1)
  app_usage_absolute = cohorts['Total_uses'].unstack(0)
  
  print(color.BOLD + '\nPercent data cohort analysis:\n')
  cohort_plot(app_usage_percent, percent=True)
  
  print(color.BOLD + '\n\nAbsolute data cohort analysis:\n')
  cohort_plot(app_usage_absolute, percent=False)
```

### Russia

```python
df_russia = df[df['Country'] == country_list[0]]
```

```python
cohort_russia = cohort_transform_plot(df_russia)
cohort_russia
```

### Egypt

```python
df_egypt = df[df['Country'] == country_list[1]]
```

```python
cohort_egypt = cohort_transform_plot(df_egypt)
cohort_egypt
```

### Botswana

```python
df_botswana = df[df['Country'] == country_list[2]]

cohort_botswana = cohort_transform_plot(df_botswana)
cohort_botswana
```

### Thailand

```python
df_thailand = df[df['Country'] == country_list[3]]

cohort_thailand = cohort_transform_plot(df_thailand)
cohort_thailand
```

In general, countries have the same individual trends as we saw in our initial analysis, with sales increasing over the months and increasing use by users in the same group.

It is worth noting that groups from some countries had a reduction in their number of users at specific times, something that had not appeared in our initial analysis. It may be interesting to look further into these specific cases to understand what may have happened and to prevent it from happening again.
